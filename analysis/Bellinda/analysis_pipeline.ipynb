## Renaming Columns

```python
clean_2015=h2015_data[["Country","Happiness Score","Health (Life Expectancy)"]]
clean_2015=clean_2015.rename(columns={"Country": "country", "Happiness Score": "2015", "Health (Life Expectancy)": "life 2015"})
clean_2015=clean_2015.sort_values(sorted(list(clean_2015[['country']])), ignore_index=True)
#print(clean_2015)       

clean_2016=h2016_data[["Country","Happiness Score","Health (Life Expectancy)"]]
clean_2016=clean_2016.rename(columns={"Country": "country", "Happiness Score": "2016", "Health (Life Expectancy)": "life 2016"})
clean_2016=clean_2016.sort_values(sorted(list(clean_2016[['country']])), ignore_index=True)

clean_2017=h2017_data[["Country","Happiness.Score","Health..Life.Expectancy."]]
clean_2017=clean_2017.rename(columns={"Country": "country", "Happiness.Score": "2017", "Health..Life.Expectancy.": "life 2017"})
clean_2017=clean_2017.sort_values(sorted(list(clean_2017[['country']])), ignore_index=True)

clean_2018=h2018_data[["Country or region","Score","Healthy life expectancy"]]
clean_2018=clean_2018.rename(columns={"Country or region": "country", "Score": "2018", "Healthy life expectancy": "life 2018"})
clean_2018=clean_2018.sort_values(sorted(list(clean_2018[['country']])), ignore_index=True)

clean_2019=h2019_data[["Country or region","Score","Healthy life expectancy"]]
clean_2019=clean_2019.rename(columns={"Country or region": "country", "Score": "2019", "Healthy life expectancy": "life 2019"})
clean_2019=clean_2019.sort_values(sorted(list(clean_2019[['country']])), ignore_index=True)
```

## Append to make one large DataFrame

```python
from functools import reduce

data_frames=[clean_2015,clean_2016,clean_2017,clean_2018,clean_2019]

hap_merged = reduce(lambda  left,right: pd.merge(left,right,on=['country'],
                                            how='inner'), data_frames)
hap_merged

#method source: https://stackoverflow.com/questions/44327999/python-pandas-merge-multiple-dataframes
```

```python

master_data=hap_merged[['country','2015','2016','2017','2018','2019']]
```

```python
just_scores=pd.melt(master_data, id_vars=['country'], value_vars=['2015', '2016','2017', '2018','2019'])
just_scores=just_scores.rename(columns={"variable":"period","value":"h_index"})
just_scores=just_scores.sort_values(sorted(list(just_scores[['country']])), ignore_index=True)
just_scores['period']=just_scores['period'].astype(int)
just_scores
```

### Functions

```python
def keep_or_discard(x):
    """Decides whether to keep or discard a piece os fruit.

    Takes in a Pandas row, and returns either 'keep' or 'discard' depending on 
    the criteria.
    """
    
    if x['period'] in [2015,2016,2017,2018,2019]:
        return 'keep'
    else:
        return 'discard'
    
def filter_years(df):
    df['status'] = df.apply(keep_or_discard,axis='columns')
    df=df.loc[df['status'] =='keep']
    df=df.drop(columns=['status'])
    return df

def edit_data(dataFrame):
    
    numbs=list(dataFrame[list(dataFrame.columns)[2]])
    for string in numbs:
        loc = string.find(' ')
        numbs[numbs.index(string)]=string[0:loc]

    return numbs
```
