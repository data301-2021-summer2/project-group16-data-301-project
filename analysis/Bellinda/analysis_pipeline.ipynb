## Renaming Columns

```python
clean_2015=h2015_data[["Country","Happiness Score","Health (Life Expectancy)"]]
clean_2015=clean_2015.rename(columns={"Country": "country", "Happiness Score": "2015", "Health (Life Expectancy)": "life 2015"})

clean_2016=h2016_data[["Country","Happiness Score","Health (Life Expectancy)"]]
clean_2016=clean_2016.rename(columns={"Country": "country", "Happiness Score": "2016", "Health (Life Expectancy)": "life 2016"})

clean_2017=h2017_data[["Country","Happiness.Score","Health..Life.Expectancy."]]
clean_2017=clean_2017.rename(columns={"Country": "country", "Happiness.Score": "2017", "Health..Life.Expectancy.": "life 2017"})

clean_2018=h2018_data[["Country or region","Score","Healthy life expectancy"]]
clean_2018=clean_2018.rename(columns={"Country or region": "country", "Score": "2018", "Healthy life expectancy": "life 2018"})

clean_2019=h2019_data[["Country or region","Score","Healthy life expectancy"]]
clean_2019=clean_2019.rename(columns={"Country or region": "country", "Score": "2019", "Healthy life expectancy": "life 2019"})
```

## Append to make one large DataFrame

```python
master_data=clean_2015.append([clean_2016,clean_2017,clean_2018,clean_2019])
print(master_data)
```

```python
just_scores=pds.melt(master_data, id_vars=['country'], value_vars=['2015', '2016','2017', '2018','2019'])
```

```python
just_scores=just_scores.rename(columns={"variable":"period","value":"h_index"})
just_scores
```

```python
clean_alco=alco[['Location','Period','Value']]
clean_homi=homi[['Location','Period','Value']]
clean_road=road[['Location','Period','Value']]
clean_suic=suic[['Location','Period','Value']]

clean_alco=clean_alco.rename(columns={"Location":"country","Period":"period","Value":"alco_con"})
clean_homi=clean_homi.rename(columns={"Location":"country","Period":"period","Value":"homi_rate"})
clean_road=clean_road.rename(columns={"Location":"country","Period":"period","Value":"road_acc"})
clean_suic=clean_suic.rename(columns={"Location":"country","Period":"period","Value":"suic_rate"})

list(clean_suic.columns)
```

```python
def keep_or_discard(x):
    """Decides whether to keep or discard a piece os fruit.

    Takes in a Pandas row, and returns either 'keep' or 'discard' depending on 
    the criteria.
    """
    
    if (x['period'] > 2014):
        return 'keep'
    else:
        return 'discard'
```

```python
indicators=[clean_alco,clean_homi,clean_road,clean_suic]

for thing in indicators:
    thing['status'] = thing.apply(keep_or_discard,axis='columns')
    thing=thing[thing['status'] =='keep']

    thing=thing[list(thing.columns)[0:3]]
    print(thing)
```
